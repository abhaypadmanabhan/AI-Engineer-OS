Single-token completion. Proves the autoregressive core.
Cost math: 8 input tok, 1 output tok on claude-sonnet-4-6 => (8/1e6)*3 + (1/1e6)*15 = $0.0000390 — effectively free.

```python
import os
from anthropic import Anthropic
assert os.environ.get('ANTHROPIC_API_KEY'), 'set ANTHROPIC_API_KEY'
client = Anthropic()
r = client.messages.create(
    model='claude-sonnet-4-6',
    max_tokens=1,
    temperature=0,
    messages=[{'role': 'user', 'content': 'The capital of France is'}],
)
print(r.content[0].text)
```
