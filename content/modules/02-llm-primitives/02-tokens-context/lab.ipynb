# Lab 02 — Tokens, context, temperature
Count tokens on 10 prompts. Compare to the chars/4 heuristic.
Cost math: count_tokens is server-side free of output tokens. 10 prompts * ~30 input tok = 300 input tok => (300/1e6)*3 = $0.0009 ~ $0.001.

```python
import os
from anthropic import Anthropic
client = Anthropic()

prompts = [
  'Summarize this 10-K in three bullets.',
  'def classify_ticket(text: str) -> Ticket: ...',
  'Translate to French: The quick brown fox jumps over the lazy dog.',
  'Write a SQL query joining users and orders on user_id.',
  'Explain transformer attention in one paragraph.',
]
for p in prompts:
    r = client.messages.count_tokens(model='claude-sonnet-4-6', messages=[{'role':'user','content':p}])
    print(r.input_tokens, 'actual vs', len(p)//4, 'guess ::', p[:40])
```

```python
# Temperature demo
def joke(t):
    r = client.messages.create(model='claude-sonnet-4-6', max_tokens=40, temperature=t,
                               messages=[{'role':'user','content':'A one-line dev joke.'}])
    return r.content[0].text
print('t=0 ->', joke(0.0))
print('t=1 ->', joke(1.0))
```
