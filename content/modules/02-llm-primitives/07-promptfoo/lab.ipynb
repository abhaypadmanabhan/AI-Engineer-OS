# Lab 07 — promptfoo matrix
Run a 4-prompt x 10-ticket bakeoff. Pick the winner.
Cost math: 4 prompts * 10 tickets * (~100 in + ~20 out) => 4000 in, 800 out => (4000/1e6)*3 + (800/1e6)*15 = $0.024 ~ $0.05 with CoT overhead.

```python
from solution import PROMPTS, TICKETS, run_prompt
from anthropic import Anthropic
client = Anthropic()
scores = {}
for name, tmpl in PROMPTS.items():
    acc = sum(run_prompt(client, tmpl, t) == lbl for t,lbl in TICKETS) / len(TICKETS)
    scores[name] = acc
    print(f'{name:12s} {acc:.2f}')
print('winner:', max(scores, key=scores.get))
```
